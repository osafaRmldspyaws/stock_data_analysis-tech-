Risk Analysis of the tech companies

```python
ret = tech_rets.dropna()
```

```python
ser1 = ret.mean()
ser2 = ret.std()

col_m = ret.mean().index
col_s = ret.std().index

ret_df = pd.DataFrame({'ret_mean':ser1,
                       'ret_std':ser2,
                       'ret_stock':col_s})
```

```python
#print(col_m == col_s)
ret_df
```

```python
plt.scatter('ret_mean','ret_std',data=ret_df)
```

```python
g = sns.factorplot('ret_mean','ret_std',data=ret_df,hue='ret_stock')
g.set_xticklabels(rotation=90)
```

Risk Analysis using bootstrap - is implemented by finding the quantiles , inadvertently finding the 95% confidence intervals for the daily_returns

```python
ret['aap_adj_close'].quantile(.05)
```
